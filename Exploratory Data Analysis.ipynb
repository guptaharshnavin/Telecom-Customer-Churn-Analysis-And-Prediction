```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

## Dataset Exploration

```python
dataset.head()
```

### Checking For Null Values

```python
dataset.isnull().sum()
```

No Null Values Are Present In The Dataset

## Exploratory Data Analysis

```python
sns.countplot(dataset['Churn'])
plt.title('Distribution Of TARGET Variable In Dataset')
plt.xlabel('Churn')
plt.ylabel('Number Of Customers')
plt.show()
```

It can be observed that the dataset, is **Highly Inbalanced**. 

```python
plt.figure(figsize=(10,6))
sns.distplot(dataset[dataset['Churn']=='No']['MonthlyCharges'], hist=True, bins=10, label='Not Churned', color='green')
sns.distplot(dataset[dataset['Churn']=='Yes']['MonthlyCharges'], hist=True, bins=10, label='Churned', color='red')
plt.title('Effect Of Monthly Charges On Churn')
plt.xlabel('Monthly Charges')
plt.ylabel('Distribution Of Customers')
plt.legend()
plt.show()
```

It can be clearly observed that, **Customers Having Monthly Charges Between 70 - 110 Have Higher Churn Distribution**.

```python
plt.figure(figsize=(10,6))
sns.countplot(dataset['PaymentMethod'], hue=dataset['Churn'], palette=['green','red'])
plt.title('Effect Of Payment Method On Customer Churn')
plt.ylabel('Number Of Customers')
plt.xlabel('Payment Method')
plt.show()
```

It can be observed that, the people who have **Electronic Check As Their Payment Method, Have Higher Churn Ratio**.

```python
plt.figure(figsize=(10,6))
sns.countplot(dataset['PaperlessBilling'], hue=dataset['Churn'], palette=['green','red'])
plt.title('Effect Of Paperless Billing On Customer Churn')
plt.ylabel('Number Of Customers')
plt.xlabel('Paperless Billing')
plt.show()
```

It can be observed that, the people who have **Paperless Billing, Have Higher Churn Ratio**.

```python
plt.figure(figsize=(10,6))
sns.distplot(dataset[dataset['Churn']=='No']['tenure'], hist=True, label='Not Churned', bins=10, color='green')
sns.distplot(dataset[dataset['Churn']=='Yes']['tenure'], hist=True, label='Churned', bins=10, color='red')
plt.legend()
plt.title('Effect Of Tenure On Customer Churn')
plt.xlabel('Tenure')
plt.ylabel('Distribution Of Customers')
plt.show()
```

It can be observed that, **Customers with a Lower Tenure, are more likely to churn than Customers with a Higher Tenure**.

```python
plt.figure(figsize=(10,6))
sns.countplot(dataset['InternetService'], hue=dataset['Churn'], palette=['green','red'])
plt.title('Effect Of Internet Service Type On Churn')
plt.xlabel('Internet Service Type')
plt.ylabel('Number Of Customers')
plt.show()
```

It can be observed that, **Customers with FibreOptic Internet Service, are more likely to churn, than other customers**.

```python
plt.figure(figsize=(10,6))
sns.countplot(dataset['TechSupport'], hue=dataset['Churn'], palette=['green','red'])
plt.title('Effect On Tech Support On Churn')
plt.xlabel('Tech Support')
plt.ylabel('Number Of Customers')
plt.show()
```

It can be seen **Customers with NO TECH SUPPORT, are more likely to churn, than customers with TECH SUPPORT**.

```python
plt.figure(figsize=(10,6))
sns.countplot(dataset['Contract'], hue=dataset['Churn'], palette=['green','red'])
plt.title('Effect Of Contract Period On Churn')
plt.xlabel('Contract Period')
plt.ylabel('Number Of Customers')
plt.show()
```

It can be seen, that **Customer Who Have Month Based Contract, have Higher Churn Ratio than other customers**.
